```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('train.csv')
data['date'] = pd.to_datetime(data['date'])
pivot_data = data.pivot(index='date', columns='warehouse', values='orders')

plt.figure(figsize=(12, 6))

for warehouse in pivot_data.columns:
    plt.plot(pivot_data.index, pivot_data[warehouse], label=warehouse)

plt.xlabel('Date')
plt.ylabel('Orders')
plt.title('Orders Over Time by Warehouse')
plt.legend(title='Warehouse')
plt.grid(True)
plt.show()


ColumnDrop= ['warehouse','date','holiday_name','id']
data.drop(columns=ColumnDrop, inplace=True)
if 'date' in data.columns:
    data['date'] = data.to_datetime(data['date'])
    data.set_index('date', inplace =True)

corr = data.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, fmt =".2f", cmap='coolwarm', cbar=True)
plt.title('Corr')
plt.show()
```
